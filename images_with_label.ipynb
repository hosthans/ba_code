```python
from training.utils import dataloader as dl
from training.utils import helper
import os
```

#### load train set

```python
data_config = helper.load_json(os.path.join("config/training", "data.json"))
dataset_config = data_config["celeba"]

train_set = dl.ImageFolder(
    config=dataset_config,
    file_path=dataset_config["train_file"],
    mode="train",
)
```

```python
selected_label = 401
selected_indices = [i for i in range (len(train_set)) if train_set.label_list[i] == selected_label]
selected_indices = selected_indices[:10]
print(selected_indices)
```

```python
import matplotlib.pyplot as plt
from torchvision import transforms
fig, axes = plt.subplots(2, 5, figsize=(10,5))

for i, idx in enumerate(selected_indices):
    img, label = train_set[idx]
    axes[i // 5, i % 5].imshow(transforms.ToPILImage()(img))  # Umwandlung in PIL Image fr die Anzeige
    axes[i // 5, i % 5].set_title(f'Label: {label}')
    axes[i // 5, i % 5].axis('off')
```
