```python
from training.utils import dataloader as dl
from training.utils import helper
import os
from torchvision import datasets, transforms
```

#### load train set

# celeba

```python
data_config = helper.load_json(os.path.join("config/training", "data.json"))
dataset_config = data_config["celeba"]

train_set = dl.ImageFolder(
    config=dataset_config,
    file_path=dataset_config["train_file"],
    mode="train",
)
```

```python
selected_label = 0
selected_indices = [i for i in range (len(train_set)) if train_set.label_list[i] == selected_label]
selected_indices = selected_indices[:10]
print(selected_indices)
```

```python
import matplotlib.pyplot as plt
from torchvision import transforms
fig, axes = plt.subplots(2, 5, figsize=(5,2))

# Entfernen von Platz zwischen Zeilen und Spalten
plt.subplots_adjust(hspace=0, wspace=0)

for i, idx in enumerate(selected_indices):
    img, label = train_set[idx]
    ax = axes[i // 5, i % 5]
    ax.imshow(transforms.ToPILImage()(img))
    # ax.set_title(f'Label: {label}')
    ax.axis('off')

# Anpassen der Gre der einzelnen Axes, um die gesamte Figur einzunehmen
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

plt.show()
```

# MNIST

```python
train_set = datasets.MNIST(
                root="datasets/mnist",
                train=True,
                download=True,
                transform=transforms.Compose(
                    [
                        transforms.Grayscale(num_output_channels=3),
                        transforms.ToTensor(),
                        transforms.Resize((64, 64)),
                        # transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
                    ]
                ),
            )
```

```python
target_label = 4

selected_indices = [i for i in range(len(train_set)) if train_set.targets[i] == target_label][:10]
selected_images = [train_set[i][0] for i in selected_indices]
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 5, figsize=(5,2))

# Entfernen von Platz zwischen Zeilen und Spalten
plt.subplots_adjust(hspace=0, wspace=0)

for i, idx in enumerate(selected_images):
    ax = axes[i // 5, i % 5]
    ax.imshow(transforms.ToPILImage()(idx))
    # ax.set_title(f'Label: {label}')
    ax.axis('off')

# Anpassen der Gre der einzelnen Axes, um die gesamte Figur einzunehmen
plt.subplots_adjust(left=0, right=1, top=1, bottom=0)

plt.show()
```
